According to the World Health Organization (WHO) stroke is the 2nd leading cause of death globally, responsible for approximately 11% of total deaths.
This dataset is used to predict whether a patient is likely to get stroke based on the input parameters.

# **Features include:-**
1) id: unique identifier

2) gender: "Male", "Female" or "Other"

3) age: age of the patient

4) hypertension: 0 if the patient doesn't have hypertension, 1 if the patient has hypertension

5) heart_disease: 0 if the patient doesn't have any heart diseases, 1 if the patient has a heart disease

6) ever_married: "No" or "Yes"

7) work_type: "children", "Govt_jov", "Never_worked", "Private" or "Self-employed"

8) Residence_type: "Rural" or "Urban"

9) avg_glucose_level: average glucose level in blood

10) bmi: body mass index

11) smoking_status: "formerly smoked", "never smoked", "smokes" or "Unknown"*

12) stroke: 1 if the patient had a stroke or 0 if not

Applying minmax scalar

```python
from sklearn.preprocessing import MinMaxScaler
```

```python

numeric_cols = ['age', 'bmi', 'avg_glucose_level']
scaler = MinMaxScaler()
scaler.fit(df[numeric_cols])
```

```python
scaled_inputs = scaler.transform(df[numeric_cols])
scaled_inputs
```

```python
scaled_inputs_df = pd.DataFrame(scaled_inputs, columns=[ i  for i in numeric_cols])
```

```python
cat_cols = ['gender', 'hypertension', 'heart_disease', 'ever_married','Residence_type','smoking_status_Unknown', 'smoking_status_formerly smoked','smoking_status_never smoked', 'smoking_status_smokes']
categorical_data = df1[cat_cols]
```

```python
scaled_df = pd.concat([scaled_inputs_df, categorical_data.reset_index(drop=True)], axis=1)
```

```python
df1.columns
```

Splitting

```python
from sklearn.model_selection import train_test_split
```

```python
x_train_,x_test_,y_train_,y_test_ = train_test_split(scaled_df,df['stroke'],test_size=0.2,random_state=101)
```

Model fitting

```python
from sklearn.linear_model import LogisticRegression
```

```python
model1=LogisticRegression()
```

```python
model1
```

```python
model1.fit(x_train_,y_train_)
```

```python
y_pred_=model1.predict(x_test_)
```

```python
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test_,y_pred_)
cm
```

```python
sns.heatmap(cm,cmap='RdPu',annot=True)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test_,y_pred_)
```
